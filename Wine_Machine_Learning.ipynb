```python
import pandas as pd

dataset = pd.read_csv('/content/wine_dataset.csv')
```

```python
dataset.head()
```

```python
#agrupa estilo do vinho.

agrupa_styles = dataset.groupby(['style']).size()
agrupa_styles
```

```python
#confere se existem valores núlos
dataset.isnull().sum()
```

```python
#altera formato de texto para número, onde o vinho vermelho é representado pelo 0
#e o vinho branco é representado pelo 1.

dataset['style'] = dataset['style'].replace('red', 0)
dataset['style'] = dataset['style'].replace('white', 1)

#verificando se foi aplicado sem erros.
dataset
```

```python
# separa as variáveis entre preditoras e variável alvo(variáveis que queremos prever).

y = dataset['style'] # variável alvo
x = dataset.drop('style', axis = 1) # recebe todo o resto das colunas (preditoras)
```

```python
# cria os conjuntos de dados de treino e teste

from sklearn.model_selection import train_test_split as tts

x_treino, x_teste, y_treino, y_teste = tts(x, y, test_size = 0.3)
```

```python
x_treino.shape
```

```python
# Cria modelo
from sklearn.ensemble import ExtraTreesClassifier as etc

model = etc()
model.fit(x_treino, y_treino)


# Imprime resultados
result = model.score(x_teste, y_teste)
print(f'Acurácia: {result}')
```
